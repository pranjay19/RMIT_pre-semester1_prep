```python
grade = 75

if grade >= 80:
    print("High Distinction (HD)")
elif grade >= 70:
    print("Distinction (DI)")
elif grade >= 60:
    print("Credit (CR)")
else:
    print("Pass (PA) or Fail (NN)")
    
# Output: Distinction (DI)
```

Applying while loop 

```python
processing_count=5

while(processing_count>0):
    print(f"processing in progress")
    processing_count=processing_count-1

print(f"processing completed !")
```

```python
student_grades = [85, "Missing", 92, 45, 78]

for grade in student_grades:
    # 1. Skip the missing data
    if grade == "Missing":
        print("Skipping missing grade...")
        continue 
        
    # 2. Stop entirely if someone failed too badly
    if grade < 50:
        print(f"Critical failure found: {grade}. Halting analysis.")
        break
        
    print(f"Processing grade: {grade}")
```

```python
# A list containing two sub-lists
university_classes = [
    ["Alex", "Jane", "Sam"], 
    ["Chris", "Pat"]
]

# The outer loop grabs the sub-lists
for classroom in university_classes:
    print("--- New Classroom ---")
    
    # The inner loop grabs the individual students
    for student in classroom:
        print(student)
```

🌟 Level-Up 1: Default Parameters

```python
# We set a default value of "Melbourne" for the campus
def enroll_student(name, campus="Melbourne"):
    return f"{name} is enrolled at the {campus} campus."

# If we don't provide a campus, it defaults to Melbourne!
print(enroll_student("Alex")) 
# Output: Alex is enrolled at the Melbourne campus.

# But we can overwrite it if we want to:
print(enroll_student("Jane", "Vietnam")) 
# Output: Jane is enrolled at the Vietnam campus.
```

🌟 Level-Up 2: Returning Multiple Things

```python
def analyze_price(raw_price):
    clean_price = float(raw_price.replace("$", ""))
    tax = clean_price * 0.10
    
    # Returning TWO things at once
    return clean_price, tax 

# You catch them on the outside using two variables
final_price, calculated_tax = analyze_price("$100.00")

print(f"Price: {final_price}, Tax: {calculated_tax}")
# Output: Price: 100.0, Tax: 10.0
```

```python
import re

messy = "Wait... is this 100% correct?!?!"

# Find everything that isn't a word/number/space, and delete it.
clean = re.sub(r"[^\w\s]", "", messy)

print(clean)
```

```python
import re

messy_string = "The price is 45 dollars and 99 cents."

# The pattern r"\d+" means: "Find any digit (\d), and grab 
# the whole cluster of them (+)"
numbers = re.findall(r"\d+", messy_string)

print(numbers) 
# Output: ['45', '99']
```
