Conditionals(if,elif,else)

```python
grade = 75

if grade >= 80:
    print("High Distinction (HD)")
elif grade >= 70:
    print("Distinction (DI)")
elif grade >= 60:
    print("Credit (CR)")
else:
    print("Pass (PA) or Fail (NN)")
    
# Output: Distinction (DI)
```

Applying For loop in List

```python
messy_data = ["  APPLE ", "bAnAnA", " CHERRY  "]
```

```python
updated_data=[]
```

```python
for i in messy_data:
    updated_data.append(i.strip().lower())
```

```python
print(updated_data)
```

Applying for loop in Dictionary

```python
student_marks = {"Alex": 85, "Jane": 92, "Sam": 45}
```

```python
for name,marks in student_marks.items():
    if marks>90:
        print(f"{name} has shown outstanding performance")
    else:
        print(f"{name} has shown good performance")

print(f"Thanks everyone for your efforts")
```

Applying while loop 

```python
processing_count=5

while(processing_count>0):
    print(f"processing in progress")
    processing_count=processing_count-1

print(f"processing completed !")
```
