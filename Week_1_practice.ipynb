1. The Cleaning Function (Day 4 & Day 5)

 - Define a function called clean_power that takes one parameter (a string like " 15.5kWh ").

 - Inside the function, use string methods to .strip() the spaces and .replace() the "kWh" with an empty string "".

 - Convert the remaining number into a float and return it.

```python
def clean_power(data):
    final_val=data.strip().replace("kWh","")
    return final_val
```

```python
for i in messy_logs:
    print(clean_power(i))
```

2. The Data Structures (Day 2)

  - Create two empty lists: valid_vehicles = [] and power_drawn = [].

```python
valid_vehicles=[]
power_drawn=[]
```

3. The Processing Loop (Day 3 & Day 5)

- Write a for loop to iterate through every string in messy_logs.

- Inside the loop, use the .split("|") method to break the log into smaller chunks.

- Write an if statement: If the string "FAILED" is anywhere in the log, use continue to skip it.

- If it is a success, extract the vehicle name, clean the punctuation, and append it to your valid_vehicles list.

- Pass the power chunk into your clean_power function, and append the returned float to your power_drawn list.

```python
array=[]

for i in messy_logs:
    array.append(i.split("|"))
    
    
```

```python
print(array)
```

```python
for i in range(len(array)):
    if "".join(re.findall(r"(FAILED)",array[i][3]))=="FAILED":
        continue
    else:
        valid_vehicles.append(re.findall(r"Vehicle:\s*(\w+)",array[i][1]))
        power_drawn.append(re.findall(r"Power:\s*(\w+)",array[i][2]))
```

```python
print(valid_vehicles)
print(power_drawn)
```

4. The Final Analysis (Day 6)

- Create a Dictionary using your two clean lists.

- Convert that Dictionary into a Pandas DataFrame.

- Run .describe() on the DataFrame to see the average power drawn across all successful sessions.

```python
dictionary_items={
    "Car": valid_vehicles,
    "Power": power_drawn
}
```
